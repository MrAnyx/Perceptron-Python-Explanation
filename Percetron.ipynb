```python
from Perceptron import Perceptron
import matplotlib.pyplot as plt
```

```python
# OR operand
inputs_training = [[0, 0], [0, 1], [1, 0], [1, 1]]
outputs_training = [0, 0, 0, 1]
nb_iter = 200
```

```python
p = Perceptron(inputs_training, outputs_training, nb_iter)
E = p.train()
```

```python
plt.plot(range(nb_iter), E)
plt.show()
```

```python
print(p.guess([0, 0]))
```
